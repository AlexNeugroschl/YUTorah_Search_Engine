```python
def convert_duration_to_seconds(duration_series: pd.Series) -> pd.Series:
        # Extract the time component from the datetime string
        time_strs = duration_series.str.split().str[1]
        # Split the time component into hours, minutes, seconds, and milliseconds
        time_parts = time_strs.str.split(':', expand=True)
        seconds_parts = time_parts[2].str.split('.', expand=True)
        time_parts[2] = seconds_parts[0]
        time_parts['milliseconds'] = seconds_parts[1]

        # Convert to total seconds
        total_seconds = (
            time_parts[0].astype(float) * 3600 +
            time_parts[1].astype(float) * 60 +
            time_parts[2].astype(float) +
            time_parts['milliseconds'].astype(float) / 1e3
        )
        return total_seconds
```

```python
series = pd.Series(['1900-01-01 00:59:49.813000'])
```

```python
convert_duration_to_seconds(series)
```

```python
def test_single_string():
    test_str = "1900-01-01 00:59:49.813000"
    time_strs = test_str.split()[1]
    time_parts = time_strs.split(':')
    seconds_parts = time_parts[2].split('.')
    hours = float(time_parts[0])
    minutes = float(time_parts[1])
    seconds = float(seconds_parts[0])
    milliseconds = float(seconds_parts[1]) if len(seconds_parts) > 1 else 0
    total_seconds = hours * 3600 + minutes * 60 + seconds + milliseconds / 1e3
    return total_seconds

```
